```python
# import libraries
import pandas as pd
import re
```

```python
# read in file
faang = pd.read_csv("data/Levels_Fyi_Salary_Data.csv")
```

```python
faang = faang[['company', 'title', 'totalyearlycompensation',
       'location', 'yearsofexperience', 'yearsatcompany','basesalary', 'bonus',
        'Masters_Degree', 'Bachelors_Degree', 'Doctorate_Degree',
       'Highschool', 'Some_College']]
```

```python
# google
company = '(Goo)+|(goo)+'
# facebookPat = '(facebook)+|(Facebook)+'
faang['google'] = faang.company.apply(lambda x: True if len(re.findall(company, str(x))) > 0 else False)
google = faang[faang.google].drop('google', axis = 1).reset_index().drop('index', axis = 1)
google['company'] = "Google"
```

```python
# save file
google.to_csv("../data/google.csv")
```
