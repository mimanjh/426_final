Read in data

```python
# import libraries
import pandas as pd
import re
```

```python
dat = pd.read_csv("data/Levels_Fyi_Salary_Data.csv")
```

Set appropriate columns

```python
faang = dat[['company', 'title', 'totalyearlycompensation',
       'location', 'yearsofexperience', 'yearsatcompany','basesalary', 'bonus',
        'Masters_Degree', 'Bachelors_Degree', 'Doctorate_Degree',
       'Highschool', 'Some_College']]
faang.head()
```

Wrangle to check all the data files has the correct company name

```python
facebookPat = '(facebook)+|(Facebook)+'
faang['facebook'] = faang.company.apply(lambda x: True if len(re.findall(facebookPat, str(x))) > 0 else False)
facebook = faang[faang.facebook].drop('facebook', axis=1).reset_index().drop('index', axis=1)
facebook['company'] = 'Facebook'
```

```python
netflixPat = '(Netflix)+|(netflix)+'
faang['netflix'] = faang.company.apply(lambda x: True if len(re.findall(netflixPat, str(x))) > 0 else False)
netflix = faang[faang.netflix].drop('netflix', axis=1).reset_index().drop('index', axis=1)
netflix['company'] = 'Netflix'
```

Export to csv

```python
facebook.to_csv('data/facebook.csv', index=False)
netflix.to_csv('data/netflix.csv', index=False)
```
